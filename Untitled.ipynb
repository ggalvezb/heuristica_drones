```python
# datos=list((demandas["Latitude"],demandas["Longitude"]))
datos=[]
for i in range(len(demandas)):
    datos.append([demandas["Latitude"][i],demandas["Longitude"][i]])   
```

```python
def clusterizacion(data,parametros_cluster):
    cl= sklearn.cluster.KMeans(n_clusters=parametros_cluster)
    clusters=cl.fit(data)
    lebels=clusters.labels_
    n_clusters = lebels.max()+1
    cluster_centers=[]
    cluster_centers= clusters.cluster_centers_
    
    return(lebels,n_clusters,cluster_centers)    
```

```python
lebels,n_clusters,cluster_centers=clusterizacion(datos,5)
print(lebels)
print(cluster_centers)
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(n_clusters), colors):
    my_members = lebels == k
    ax.scatter(demandas_original[my_members]['Latitude'], demandas_original[my_members]['Longitude'], c=col, marker='.')
plt.show()
```

```python
clusters={}
#Creo el diccionario cluster con las llaves y listas vacias
for i in range(n_clusters):
    key='cluster_'+str(i+1)
    clusters[key]=[] 

#lleno el diccionario con los nodos pertenecientes de cada cluster
for i in range(len(lebels)):
#     clusters['cluster_'+str(lebels[i]+1)].append(demandas.loc[i]['id'])
    clusters['cluster_'+str(lebels[i]+1)].append(demandas.loc[i])

print(clusters)    
```
